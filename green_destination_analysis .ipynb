```python
#Visualizations with seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

# Age vs Attrition

```python
#Age vs Attrition (Histogram plot)
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Age',hue='Attrition',multiple='stack',color='green',edgecolor='orange',alpha=0.5,bins=30)
plt.title('Age vs Attrition')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
#
```

```python
sns.boxplot(x='Attrition',y='Age',data=df)
plt.title('age vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.show()
```

```python
df.groupby('Attrition')['Age'].mean()
```

#Observation :Age vs Attrition
The plot shows that the employees who left are slightly younger (28-39 yrs) compared to those stayed(32-43 years).However, since there's significant overlap,age alone doesnt appear to strongly influence attrition.

The average age of employees who stayed at the comapny is 37.56 and who left is 33.6.The younger employees are more likely to leave compared to older, while the difference is not very large.Therefor age have a weak to moderate impact on attrition.

# Monthly income Vs attrition

```python
# Monthly income vs attrition(Histogram)
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='MonthlyIncome',hue='Attrition',multiple='stack',color='green',edgecolor='orange',alpha=0.5,bins=30)
plt.title('Monthly Income vs Attrition')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.show()
```

```python
#Monthly income vs attrition(Boxplot)
sns.boxplot(x='Attrition',y='MonthlyIncome',data=df)
plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

```python
df.groupby('Attrition')['MonthlyIncome'].mean()
```

#Observation: Monthly income vs Attrition
The box plot shows some noticable difference between those left and those stayed.The IQR for employees who left the company earned between 2500 and 6000,while those who stayed had a IQR of 3000 and 9000.suggesting that higher earners are more likely to remain with the company suggests that income has a moderate to strong relationship with attrition.Because outliers in both groups indicates that a very high earners also left or stayed.

The average monthly income of employees who stayed at the company is 6832.7 and who left the company is 4787.1.This clear difference suggests that low income employees are likely to leave.

#Years at company vs attrition

```python
# Years at company vs attrition(Box plot)
sns.boxplot(x='Attrition',y='YearsAtCompany',data=df)
plt.title('Years at Company vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years at Company')
plt.show()
```

```python
#Average year at company
df.groupby('Attrition')['YearsAtCompany'].mean()
```

# Observation Years at company Vs Attrition
The box plot shows that the employees who left typically had less experience at company (1 to 7.5 yrs). On other hand those who stayed had a higher IQR of 2.5 to 10 yrs.This suggests that newer employees are more likely to leave. While those with longer tenure tend to stay.There is still some overlap in the 2.5 to 7.5 year range, indicating that tenure alone doesnt fully determine attrition,but it does have moderate influence.
Outliers in both the groups-Emloyees with over 20 years at the company -are rare and do not effect the overall trend.

The Average years at the company is 5.13 years for those left and 7.37 for those who stayed.This also supports the idea that newer employees are more likely to leave.
